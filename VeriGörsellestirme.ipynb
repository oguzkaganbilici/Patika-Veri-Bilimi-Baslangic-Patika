```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
x = np.arange(-10,11)
x
```

```python
plt.figure(figsize=(12,6))

plt.title("My Nice Plot")

plt.plot(x, x**2)
plt.plot(x, -1 * (x**2))
```

```python
plt.figure(figsize=(12,6))
plt.title("My Better Plot")

plt.subplot(1,2,1) #rows, columns, panel selected
plt.plot(x, x**2)
plt.plot([0,0,0], [-10, 0, 100])
plt.legend(["X^2", "Vertical Line"])
plt.xlabel("X")
plt.ylabel("X Squarred") 

plt.subplot(1,2,2)
plt.plot(x, -1 * (x**2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(["-X^2", "Horizontal Line"])
plt.xlabel("X^2")
plt.ylabel("X Squarred")
```

- Bunu daha farkl sekilde de gerceklestirebiliriz.

```python
fig, axes = plt.subplots(figsize=(12,6))
axes.plot(
    x, (x**2), color="red", linewidth=3, marker="o", markersize=8, label="X^2")

axes.plot(
    x, -1 *(x**2), "b--", label="-X^2")

axes.set_xlabel("X")
axes.set_ylabel("X Squared")

axes.set_title("My the best title")

axes.legend()
```

- tek plot zerinde farkl farkli grafikler izdirebiliriz

```python
fig, axes = plt.subplots(figsize=(12,6))

axes.plot(x, x + 0, "-og", label="solid green")
axes.plot(x, x + 1, "--c", label="dashed cyan")
axes.plot(x, x + 2, "-.b", label="dashdot blue")
axes.plot(x, x + 3, ":r", label="dotted red")

axes.set_title("My amazing plot")

axes.legend()
```

### Scatter Plot

```python
N = 50
x = np.random.randn(N)
y = np.random.randn(N)
colors = np.random.randn(N)
area = np.pi * (20 * np.random.randn(N)) ** 2
```

```python
plt.figure(figsize=(14,6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap="Spectral")
plt.colorbar()

plt.show()
```

```python
fig = plt.figure(figsize=(14,6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap="Pastel1")
plt.colorbar()

ax2=fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=.5, cmap="Pastel2")
plt.colorbar()

plt.show()
```

### Histograms

```python
values = np.random.randn(1000)
```

```python
plt.subplots(figsize=(12,6))

plt.hist(values, bins=100, alpha=.8, histtype="bar", color="steelblue", edgecolor="green")
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### KDE (Kernel Density Estimation)

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12,6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)
plt.plot(values2, density(values2), color="#FF7F00")
plt.fill_between(values2, density(values2), alpha=.5, color="blue")
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### Combine Plots

```python
plt.subplots(figsize=(12,6))

plt.hist(values, bins=100, alpha=.8, density=1, histtype="bar", color="steelblue", edgecolor="green")

plt.plot(values2, density(values2), color="#FF7F00", linewidth=3)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### Bar plots

```python
Y = np.random.randn(1,5)[0]
Y2 = np.random.randn(1,5)[0]
print("Y: ", Y)
print("Y2: ", Y2)
```

```python
plt.figure(figsize=(12,4))

barWidth=.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color="#00b894")

plt.show()
```

```python
plt.figure(figsize=(12,4))

barWidth = .5

plt.bar(np.arange(len(Y)), Y, width=barWidth, color="#00b894", label="Label Y")
plt.bar(np.arange(len(Y)), Y2, width=barWidth, color="#e17055", bottom=Y, label="Label Y2")

plt.legend()
plt.show()
```
